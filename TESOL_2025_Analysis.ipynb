### a. Mapping the Background Information variables

```python
# Removing unnecessary columns based on the correct column names
mrh_data_cleaned = mrh_data.drop(columns=['ip_address', 'time_spent_second'])

# Simplify long categorical values
simplification_mappings_bkgrinfo = {
    # Teaching Level
    'teaching_level': {
        'Other (Please specify: ____________________)': 'other',
        'I teach English in a middle school': 'middle school',
        'I teach English in a secondary school': 'high school',
        'I teach English in an elementary school': 'elementary school'
    },
    # Teacher Type
     'teacher_type': {
        'I am an-service teacher (i.e., I am currently working full-time as a teacher)': 'in-service full-time',
        'I am a pre-service teacher (i.e., I am currently in training at ​Centre régional des métiers de l’éducation et de la formation (CRMEF) or Ecole Normale Supérieure (ENS)': 'pre-service in-training',
        'Other (please specify: _______________________)': 'other'
    },
    # School type
    'school_type': {
       'Private school' : 'private',
       'Public school' : 'public',
       'Not currently teaching' : 'not teaching'
    },
    'school_location': {
        'Rural': 'rural',
        'Urban': 'urban',
        'Suburban': 'suburban',
        'Click to write Choice 4': 'other'
    }
}

# Apply the mappings to the relevant columns
for column, mapping in simplification_mappings_bkgrinfo.items():
    mrh_data_cleaned[column] = mrh_data_cleaned[column].replace(mapping)

# Verify the changes
mrh_data_cleaned['teaching_level'].value_counts()
```

```python
# Summaries for each background variable
# Define the remaining background variables
background_variables = [
    'gender', 'age_range', 'teaching_experience', 'school_location',
    'teacher_type', 'school_type', 'teaching_level', 'school_region'
]

# Function to calculate frequency and percentage distribution
def summarize_categorical(data, columns):
    summaries = {}
    for column in columns:
        frequency = data[column].value_counts()
        percentage = data[column].value_counts(normalize=True) * 100
        summary = pd.DataFrame({'Frequency': frequency, 'Percentage': percentage})
        summaries[column] = summary
    return summaries

# Generate summaries for background variables
background_summaries = summarize_categorical(mrh_data_cleaned, background_variables)

# Display the summaries
background_summaries
```

### Visualizing the Background Information Appropirately
Using Bar Plot, Pie Chart, and Count Plots.
1. *gender*: Pie Chart to show the proportion of genders.
2. *age_range*: Bar Plot to display the distribution of age ranges.
3. *teaching_experience*: Bar Plot to show the distribution of teaching experience.
4. *school_location*: Pie Chart to show the proportion of school locations.
5. *teacher_type*: Bar Plot to display the distribution of teacher types.
6. *school_type*: Bar Plot to show the distribution of school types.
7. *teaching_level*: Bar Plot to display the distribution of teaching levels.
8. *school_region*: Bar Plot to show the distribution of school regions.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plots
def plot_and_save_categorical_data(summaries):
    for variable, summary in summaries.items():
        plt.figure(figsize=(10, 6))

        if variable == 'gender':
            # Pie Chart for gender
            plt.pie(summary['Frequency'], labels=summary.index, autopct='%1.1f%%', startangle=140)
            plt.title(f'Pie Chart of {variable}')
            plt.savefig(f'{variable}_pie_chart.png')
            plt.show()

        elif variable == 'age_range':
            # Bar Plot for age_range
            sns.barplot(x=summary.index, y='Frequency', data=summary)
            plt.title(f'Bar Plot of {variable}')
            plt.xlabel(variable)
            plt.ylabel('Frequency')
            plt.xticks(rotation=45)
            plt.savefig(f'{variable}_bar_plot.png')
            plt.show()

        elif variable == 'teaching_experience':
            # Bar Plot for teaching_experience
            sns.barplot(x=summary.index, y='Frequency', data=summary)
            plt.title(f'Bar Plot of {variable}')
            plt.xlabel(variable)
            plt.ylabel('Frequency')
            plt.xticks(rotation=45)
            plt.savefig(f'{variable}_bar_plot.png')
            plt.show()

        elif variable == 'school_location':
            # Pie Chart for school_location
            plt.pie(summary['Frequency'], labels=summary.index, autopct='%1.1f%%', startangle=140)
            plt.title(f'Pie Chart of {variable}')
            plt.savefig(f'{variable}_pie_chart.png')
            plt.show()

        elif variable == 'teacher_type':
            # Bar Plot for teacher_type
            sns.barplot(x=summary.index, y='Frequency', data=summary)
            plt.title(f'Bar Plot of {variable}')
            plt.xlabel(variable)
            plt.ylabel('Frequency')
            plt.xticks(rotation=45)
            plt.savefig(f'{variable}_bar_plot.png')
            plt.show()

        elif variable == 'school_type':
            # Bar Plot for school_type
            sns.barplot(x=summary.index, y='Frequency', data=summary)
            plt.title(f'Bar Plot of {variable}')
            plt.xlabel(variable)
            plt.ylabel('Frequency')
            plt.xticks(rotation=45)
            plt.savefig(f'{variable}_bar_plot.png')
            plt.show()

        elif variable == 'teaching_level':
            # Bar Plot for teaching_level
            sns.barplot(x=summary.index, y='Frequency', data=summary)
            plt.title(f'Bar Plot of {variable}')
            plt.xlabel(variable)
            plt.ylabel('Frequency')
            plt.xticks(rotation=45)
            plt.savefig(f'{variable}_bar_plot.png')
            plt.show()

        elif variable == 'school_region':
            # Bar Plot for school_region
            sns.barplot(x=summary.index, y='Frequency', data=summary)
            plt.title(f'Bar Plot of {variable}')
            plt.xlabel(variable)
            plt.ylabel('Frequency')
            plt.xticks(rotation=45)
            plt.savefig(f'{variable}_bar_plot.png')
            plt.show()

# Plot and save the data
plot_and_save_categorical_data(background_summaries)
```

### b. Mapping the Digital Device Access Variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# List of specific categorical variables
device_variables = [
    'decide_cellphone', 'device_smartphone_winternet', 'device_mp3_ipod',
    'device_laptop_tablet', 'device_desktop', 'device_hspeed_internet',
    'device_tv', 'device_dvd_cd', 'device_vidgame_player'
]

# Function to calculate frequency and percentage distribution for specific device variables
def summarize_devices(data, columns):
    summaries = {}
    for column in columns:
        frequency = data[column].value_counts()
        percentage = data[column].value_counts(normalize=True) * 100
        summary = pd.DataFrame({'Frequency': frequency, 'Percentage': percentage})
        summaries[column] = summary
    return summaries

# Generate summaries for device variables
device_summaries = summarize_devices(mrh_data_cleaned, device_variables)

# Display the summaries
for device, summary in device_summaries.items():
    print(f"Summary for {device}:\n")
    display(summary)
    print("\n")

# Function to convert summaries into a single DataFrame for plotting
def create_summary_dataframe(summaries):
    summary_list = []
    for device, summary in summaries.items():
        summary['Device'] = device
        summary['Response'] = summary.index
        summary_list.append(summary)
    return pd.concat(summary_list)

# Create a combined summary DataFrame
combined_summary_df = create_summary_dataframe(device_summaries)

# Plotting the stacked bar plot
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")
palette = sns.color_palette("muted")

# Create the plot
barplot = sns.barplot(x='Device', y='Frequency', hue='Response', data=combined_summary_df, palette=palette, dodge=False)

# Add title and labels with APA style
plt.title('Device Ownership and Access', fontsize=14, weight='bold')
plt.xlabel('Device', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# Adding data labels
for p in barplot.patches:
    height = p.get_height()
    barplot.annotate(f'{height:.0f}',
                     (p.get_x() + p.get_width() / 2., height),
                     ha='center', va='center', xytext=(0, 5),
                     textcoords='offset points', fontsize=10)

# Customize legend
plt.legend(title='Response', title_fontsize='12', fontsize='10', loc='upper right')

# Save and display the plot
plt.tight_layout()
plt.savefig('device_ownership_access.png')
plt.show()
```
